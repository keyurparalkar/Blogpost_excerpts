## Method 1: Reading the data by changing the datatypes

```python
#Creating low-end data type list:
dtype_list = list()
for x in sample_train.dtypes.tolist():
    if x=='int64':
        dtype_list.append('int16')
    elif(x=='float64'):
        dtype_list.append('float16')
    else:
        dtype_list.append('object')
        
dtype_list = dict(zip(sample_train.columns.tolist(),dtype_list))
dtype_list
```

```python
%time pd.read_csv('train_V2.csv',dtype=dtype_list)
```

```python
%memit pd.read_csv('train_V2.csv',dtype=dtype_list)
```

## Method 3: Reading the Data in the form of chunks

```python
#defining chunk-size:
chunk_size = 10**6
chunk_size
```

```python
%timeit [chunk for chunk in pd.read_csv('train_V2.csv',chunksize=chunk_size)]
```

```python
%memit [chunk for chunk in pd.read_csv('train_V2.csv',chunksize=chunk_size)]
```

```python
data_chunks = [chunk for chunk in pd.read_csv('train_V2.csv',chunksize=chunk_size)]
len(data_chunks)
```
